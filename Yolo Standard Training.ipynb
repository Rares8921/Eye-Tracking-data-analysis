```python
import os
import cv2
import numpy as np
from ultralytics import YOLO
import platform
import ctypes # pt Windows
```

```python
model = YOLO("yolo11s.pt")

yaml_path = "dataset/waldo.yaml"
#https://docs.ultralytics.com/modes/train/
model.train(
    data=yaml_path,
    epochs=50,
    workers=2,
    patience=10,
    imgsz=1280,
    batch=2,
    name="yolo_standard",
    device=0,
    optimizer='AdamW',
    lr0=0.001,
    cos_lr=True,
    scale=0.5,
    fliplr=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
)
```

```python
metrics = model.val(device=0)
metrics
```

```python
model = YOLO("runs/detect/yolo_standard/weights/best.pt").to("cuda")

def preprocess_fast(img):
    k = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]], dtype=np.float32)
    img = cv2.filter2D(img, -1, k)
    img = cv2.convertScaleAbs(img, alpha=1.2, beta=0)
    return img

def detect_and_show(img_path):
    img = cv2.imread(img_path)
    if img is None: return

    prep = preprocess_fast(img)
    window_name = os.path.basename(img_path)

    results = model.predict(prep, verbose=False, imgsz=1280, conf=0.25, max_det=1)
    found = len(results[0].boxes) > 0

    if not found:
        results = model.predict(prep, verbose=False, imgsz=1920, conf=0.15, augment=True, max_det=1)

    for box, conf in zip(results[0].boxes.xyxy.cpu().numpy(),
                         results[0].boxes.conf.cpu().numpy()):
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(prep, (x1, y1), (x2, y2), (0, 255, 0), 4)

        label = f"Waldo {conf:.2f}"
        t_size = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.8, 2)[0]
        cv2.rectangle(prep, (x1, y1 - t_size[1] - 5), (x1 + t_size[0], y1), (0, 255, 0), -1)
        cv2.putText(prep, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 0), 2)

    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)

    cv2.resizeWindow(window_name, 1600, 900)
    cv2.moveWindow(window_name, 0, 0)

    cv2.imshow(window_name, prep)

    if platform.system() == "Windows":
        try:
            user32 = ctypes.windll.user32
            hwnd = user32.FindWindowW(None, window_name)
            if hwnd:
                user32.ShowWindow(hwnd, 3) # 3 = SW_MAXIMIZE
        except Exception as e:
            print("fail resize")

    cv2.waitKey(0)
    cv2.destroyAllWindows()

paths = [f"dataset/test/images/level{i}.png" for i in range(1, 5)]
for p in paths:
    detect_and_show(p)
```
